```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
data = pd.read_csv('AllCasesTimeQualityAvg1.csv')
# Create scatter plot
plt.figure(figsize=(15, 6))
plt.scatter(data['MaxTime'], data['Quality'], c=data['Color'])
# Adding regression line



data = pd.read_csv('AllCasesTimeQualityAvg1.csv')
filtered_data_1 = data[data['Algorithm'] == 'Greedy']
filtered_data_2 = data[data['Algorithm'] == 'MinCost']
filtered_data_3 = data[data['Algorithm'] == 'AntColony']
filtered_data_4 = data[data['Algorithm'] == 'AntGreedy']
filtered_data_5 = data[data['Algorithm'] == 'AntMinCost']
filtered_data_6 = data[data['Algorithm'] == 'SimulatedAnnealingGreedy']
filtered_data_7 = data[data['Algorithm'] == 'SimulatedAnnealingMinCost']
filtered_data_8 = data[data['Algorithm'] == 'SimulatedAnnealingAnt']
filtered_data_9 = data[data['Algorithm'] == 'SimulatedAnnealingEmpty']
filtered_data_10 = data[data['Algorithm'] == 'SimulatedAnnealingFullyRandom']

plt.plot(filtered_data_1['MaxTime'], np.poly1d(np.polyfit(filtered_data_1['MaxTime'], filtered_data_1['Quality'], 1))(filtered_data_1['MaxTime']), label= filtered_data_1['Algorithm'][0], color='blue')
plt.plot(filtered_data_2['MaxTime'], np.poly1d(np.polyfit(filtered_data_2['MaxTime'], filtered_data_2['Quality'], 1))(filtered_data_2['MaxTime']), label= filtered_data_2['Algorithm'][7], color='red')
plt.plot(filtered_data_3['MaxTime'], np.poly1d(np.polyfit(filtered_data_3['MaxTime'], filtered_data_3['Quality'], 1))(filtered_data_3['MaxTime']), label= filtered_data_3['Algorithm'][14], color='green')
plt.plot(filtered_data_4['MaxTime'], np.poly1d(np.polyfit(filtered_data_4['MaxTime'], filtered_data_4['Quality'], 1))(filtered_data_4['MaxTime']), label= filtered_data_4['Algorithm'][21], color='orange')
plt.plot(filtered_data_5['MaxTime'], np.poly1d(np.polyfit(filtered_data_5['MaxTime'], filtered_data_5['Quality'], 1))(filtered_data_5['MaxTime']), label= filtered_data_5['Algorithm'][28], color='purple')
plt.plot(filtered_data_6['MaxTime'], np.poly1d(np.polyfit(filtered_data_6['MaxTime'], filtered_data_6['Quality'], 1))(filtered_data_6['MaxTime']), label= filtered_data_6['Algorithm'][35], color='cyan')
plt.plot(filtered_data_7['MaxTime'], np.poly1d(np.polyfit(filtered_data_7['MaxTime'], filtered_data_7['Quality'], 1))(filtered_data_7['MaxTime']), label= filtered_data_7['Algorithm'][42], color='magenta')
plt.plot(filtered_data_8['MaxTime'], np.poly1d(np.polyfit(filtered_data_8['MaxTime'], filtered_data_8['Quality'], 1))(filtered_data_8['MaxTime']), label= filtered_data_8['Algorithm'][49], color='black')
plt.plot(filtered_data_9['MaxTime'], np.poly1d(np.polyfit(filtered_data_9['MaxTime'], filtered_data_9['Quality'], 1))(filtered_data_9['MaxTime']), label= filtered_data_9['Algorithm'][56], color='brown')
plt.plot(filtered_data_10['MaxTime'], np.poly1d(np.polyfit(filtered_data_10['MaxTime'], filtered_data_10['Quality'], 1))(filtered_data_10['MaxTime']), label= filtered_data_10['Algorithm'][63], color='dodgerblue')

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.title('Ant Colony Cases')
plt.xlabel('MaxTime')
plt.ylabel('Quality')
plt.grid(True)
plt.show()
```
